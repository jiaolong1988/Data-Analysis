## 标尺转换

```python
# 默认是将输数字转换为对数
def log_trans(x, inverse = False):
    if not inverse:
        #对数转换
        return np.log10(x)
    else:
        #对数还原
        return np.power(10, x)
```

```python
#数据缩放展示
sb.regplot(fuel_econ['displ'], fuel_econ['comb'].apply(log_trans));
tick_locs = [10, 20, 30, 40, 50, 60];

plt.yticks(log_trans(tick_locs), tick_locs);#参数：刻度位置，刻度
```

## plt.hist2d

```python
plt.figure(figsize = [12, 5])

plt.subplot(1, 2, 1)
#散点图
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', 
           fit_reg = False,                 # fit_reg = False 表示取消回归线显示
           x_jitter = 0.2, y_jitter = 0.2,  # x_jitter/y_jitter = 0.2 设置抖动将导致每个点统一地在真实值的 ±0.2 范围内抖动
           scatter_kws = {'alpha' : 1/3})   # 'alpha'的值可以介于 0（完全不透明，不可见） 到 1（完全透明）之间

plt.subplot(1, 2, 2)

#自定义组距：根据describe()设定
bins_x = np.arange(0.6, 7+0.3, 0.3)
bins_y = np.arange(12, 58+3, 3)

#热图
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', 
           cmin = 0.5 ,                   # 指定统计值下限，只有大于 统计值下限 才会得到颜色填充。
           cmap='viridis_r',              # 将调色板颜色翻转展示
           bins = [bins_x, bins_y]        # 设置组距：顺序指明 x 轴和 y 轴的组边界
          );

#添加色条，显示从计数到颜色的映射
plt.colorbar(); 

plt.xlabel('displ');
plt.ylabel('comb');
```

## 添加标记

```python
bins_x = np.arange(0.6, 7+0.3, 0.3)
bins_y = np.arange(12, 58+3, 3)
h2d = plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', 
           cmin = 0.5 ,      # 指定统计值下限，只有大于 统计值下限 才会得到颜色填充。
           cmap='viridis_r',  # 将调色板颜色翻转展示
           bins = [bins_x, bins_y] #顺序指明 x 轴和 y 轴的组边界
          );
plt.colorbar(); #添加色条，显示从计数到颜色的映射
counts = h2d[0]

# loop through the cell counts and add text annotations for each
# 通常只会在单元格很少的分类热图中添加注释
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 7: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'black')
```

# 小提琴图-sb.violinplot

'Minicompact Cars':迷你型紧凑级车  
 'Subcompact Cars':次紧凑级车  
'Compact Cars':小型车  
 'Midsize Cars':中型车  
'Large Cars':大型车

```python
#对VClass字段值设定显示顺序
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', 
              color = base_color
             #,inner = None #不显示小提琴图中箱线图
             );
# X轴字段旋转15度显示
plt.xticks(rotation=15);
```

由图可知：
1. 汽车大小与燃油效率的关系并不单一，虽然最大的汽车燃油效率最低，但这条长尾巴意味着该车型在最高效率处有异常值。
2. 微型车和小型车的燃油效率为第二高，但微型车的分布有更多峰值
3. 最后紧凑型车和中型车的燃油效率最高但中型车的分布要比紧凑型的广

```python
#inner = 'quartile' 虚线显示小提琴图中的 箱线图
sb.violinplot(data = fuel_econ,x = 'VClass', y = 'comb', color = base_color, inner = 'quartile');
plt.xticks(rotation=15);
```

# 分面-FacetGrid

```python
'''
    对字段VClass字段进行分面，对字段comb进行直方图频率统计。
    hist默认划分10个分箱数据
'''
bins = np.arange(12, 58+2, 2) #生成12-60范围的数字，每个数字间隔为2
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap = 3) # col_wrap=3 设置并排放置的分面数量
g.map(plt.hist, 'comb', bins=bins);# map方法第一个参数指定绘制的图形，
```

```python
bins = np.arange(12, 58+2, 2) 
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap=3, sharey=False) # sharey=False 设置每个分面使用不同Y轴范围
g.map(plt.hist, 'comb', bins=bins);
```

```python
# 设置指定分面类别顺序
group_means = fuel_econ.groupby(['trans']).mean() # 对字段trans分组，并对其他数据求平均值
group_order = group_means.sort_values(['comb'], ascending = False).index # 对数据排序并获得索引，也就是获取trans在指定条件下的排列顺序。

g = sb.FacetGrid(data = fuel_econ, col = 'trans', col_wrap = 5,
                 height = 2 , # 设置图片的高度
                 col_order = group_order # 指定分面类别的顺序
                )

g.map(plt.hist, 'comb', bins = np.arange(12, 58+2, 2))
g.set_titles('{col_name}'); # 设置标题名称
```

# 调整单变量图形-sb.barplot

```python
fuel_econ.groupby('VClass').mean()['comb']
```

```python
'''
    1. 单变量下多类别，统计 每个类别下 指标指标 的平均值。
    2. 代码功能：对 VClass字段下每个类别的comb字段 求平均值。
    3. countplot簇状柱状图的升级，见8节
'''
#柱形上的黑线表示平均值的不确定程度 供我们参考
base_color = sb.color_palette()[0]
sb.barplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color) # Y轴是comb的平均值
plt.xticks(rotation=15);
plt.ylabel('comb的平均值');
```

```python
#柱形上的黑线表示平均值的不确定程度 供我们参考
base_color = sb.color_palette()[0]
sb.barplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color,
           #errwidth = 0 , #不显示误差线
           ci = 'sd' # 显示数据点的标准差  
          ) 
plt.xticks(rotation=15);
plt.ylabel('comb的平均值');
```

```python
#如果你不想看到为 0 的下限，你可以将函数 barplot 换成 pointplot
sb.pointplot(data = fuel_econ, x = 'VClass', y = 'comb',
           #errwidth = 0 ,       #不显示误差线
           ci = 'sd'            # 显示数据点的标准差  
           #, linestyles = ""   #当ci = 'sd'时，设置该参数可将中间的连接线去除。
          ) 
plt.xticks(rotation=15);
plt.ylabel('comb的平均值');
```

# 折线图-plt.errorbar

分析燃油效率和发动机排量之间的关系

```python
plt.errorbar(data= fuel_econ, x ='displ', y = 'comb');
```

```python
# 设置组边界和中心，对排量数据进行汇总，我要用中心值来准确绘制数据点
bins_e = np.arange(0.6, 7+0.2, 0.2)
bins_c = bins_e[:-1] + 0.1
```

```python
# 确定各数据点对应哪一组：
# 第一个参数是一个切片出来的序列，
# 第二个参数是组集合，
# 第三个参数用于确保等于平均值的值
displ_binned = pd.cut(fuel_econ['displ'], bins_e, include_lowest = True) 

#给排量组分组 然后取每组数据点的平均值
comb_mean = fuel_econ['comb'].groupby(displ_binned).mean()
comb_std = fuel_econ['comb'].groupby(displ_binned).std()
```

```python
#参数 x 设置组中心
#参数 y 是燃油效率Pandas序列
plt.errorbar(x = bins_c, y = np.array(comb_mean));
```

```python
# yerr 设置标准差
plt.errorbar(x = bins_c, y = np.array(comb_mean), yerr = np.array(comb_std));
```

- 基本上发动机越大 燃油效率越低
- 函数 pointplot擅长总结定量变量与定性变量之间的关系
- 两个定量变量的关系，要用 Matplotlib 的函数 errorbar
